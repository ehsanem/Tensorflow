#Theory of CNN
Instead of Dense networks that all neurons of layers are connect to each other, in convolutional network each neuron summerizes input data based on a certain criteria. It behaives like a student who takes note from lecture slides.

**(0) Layer 1**
The first layer is the input of network with specified size.

**(1) Convolutional Layer**
In the second layer, we define how this layer (convolution layer) summerize the input layer.

**(2) Pooling Layer**
summerize the ourtput of filters and usually takes max or mean of the filter output.

**(3) Last Layers**
Finally, output of the last convolution layer will be the input of a series of dense layer in order to apply classification and decide what is the class of the input.

image.png

##Term definitions
1. **Filter**: a series of filters that applying on input data in order to extract features.
2. **Filter map**: the output of filter called filter map
3. **strides**: pixel step that neuron takes to scan next part of the input data.
4. **padding**: adding or removing margins from input data to make it convenient for neuron scan.
